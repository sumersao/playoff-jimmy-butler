```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
all_stars = pd.read_csv('data/all_stars_averages.csv')
adv_stats_reg = pd.read_csv('data/advanced_stats.csv')
adv_stats_playoffs = pd.read_csv('data/advanced_stats_playoffs.csv')

all_star_names = all_stars['PLAYER_NAME'].unique()

adv_stats_reg = adv_stats_reg.loc[adv_stats_reg['Player'].isin(all_star_names)]
adv_stats_playoffs = adv_stats_playoffs.loc[adv_stats_playoffs['Player'].isin(all_star_names)]

# jimmy butler was on 2 teams in 2019 so there are 2 regular season rows. Drop the second one
adv_stats_reg = adv_stats_reg.drop(adv_stats_reg.index[(adv_stats_reg['Player'] == 'Jimmy Butler') & (adv_stats_reg['Year'] == 2019)].tolist()[1])
```

```python
# within each year, rank the players by BPM
adv_stats_reg['BPM_Rank'] = adv_stats_reg.groupby('Year')['BPM'].rank(ascending=False)
adv_stats_playoffs['BPM_Rank'] = adv_stats_playoffs.groupby('Year')['BPM'].rank(ascending=False)

# plot jimmy butler's BPM rank over time
jimmy_butler = adv_stats_reg.loc[adv_stats_reg['Player'] == 'Jimmy Butler']
jimmy_butler_playoffs = adv_stats_playoffs.loc[adv_stats_playoffs['Player'] == 'Jimmy Butler']

jimmy_butler = jimmy_butler.sort_values(by='Year')
jimmy_butler_playoffs = jimmy_butler_playoffs.sort_values(by='Year')

# drop the 4th row of the jimmy_butler dataframe
jimmy_butler = jimmy_butler.drop(jimmy_butler.index[3])

# label each point with the BPM rank
for i in range(len(jimmy_butler)):
    plt.annotate(jimmy_butler['BPM_Rank'].iloc[i], (jimmy_butler['Year'].iloc[i], jimmy_butler['BPM_Rank'].iloc[i]))

for i in range(len(jimmy_butler_playoffs)):
    plt.annotate(jimmy_butler_playoffs['BPM_Rank'].iloc[i], (jimmy_butler_playoffs['Year'].iloc[i], jimmy_butler_playoffs['BPM_Rank'].iloc[i]))

plt.plot(jimmy_butler['Year'], jimmy_butler['BPM_Rank'], label='Regular Season')
plt.plot(jimmy_butler_playoffs['Year'], jimmy_butler_playoffs['BPM_Rank'], label='Playoffs')
plt.legend()
plt.xlabel('Year')
plt.ylabel('BPM Rank')
plt.title('Jimmy Butler BPM Rank Over Time')
plt.savefig('plots/jimmy_butler_bpm_rank.png')
plt.show()
```
